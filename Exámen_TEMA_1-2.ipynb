#### 1. ¿Cuál es el promedio de la cantidad de veces que debo de jugar hasta que logre ganar (tres carretes iguales)? ¿Cuál es su desviación estándar? (dibuja su histograma).

```python
def simular_juego(_):
    juegos = 0
    while True:
        juegos += 1
        carretes = np.random.randint(1, 11, size=3)
        if len(set(carretes)) == 1:
            return juegos

op_para_ganar = list(map(simular_juego, range(num_simulaciones)))
promedio = np.mean(op_para_ganar)
desviacion_estandar = np.std(op_para_ganar)

print("Promedio de veces para ganar:", promedio)
print("Desviación estándar de veces para ganar:", desviacion_estandar)

plt.hist(op_para_ganar, bins=range(1, max(op_para_ganar)+1), edgecolor='black')
plt.xlabel('Oportunidad de veces para ganar')
plt.ylabel('Frecuencia')
plt.title('Histograma de la cantidad de veces para ganar')
plt.show()
```

#### 2. Para el inciso anterior ¿Cuál es la probabilidad de que juegues más de 500 veces hasta que ganes un juego? ¿ y de que juegues menos de 400 veces?

```python
juegos_mas_de_500 = 0
juegos_menos_de_400 = 0

for juegos in op_para_ganar:
    if juegos > 500:
        juegos_mas_de_500 += 1

for juegos in op_para_ganar:
    if juegos < 400:
        juegos_menos_de_400 += 1
        
prob_mas_de_500 = juegos_mas_de_500 / num_simulaciones
prob_menos_de_400 = juegos_menos_de_400 / num_simulaciones

print("la probabilidad de que juegues más de 500 veces hasta que ganes un juego:", prob_mas_de_500)
print("La probabilidad de  juegues menos de 400 veces:", prob_menos_de_400)
```

#### 3. ¿Cuál es la probabilidad de ganar si un jugador juega 100 veces en la máquina tragamonedas?

```python
cien_juegos = 0

for juegos in op_para_ganar:
    if juegos == 100:
        cien_juegos += 1
    
probabilidad_de_ganar_cien_juegos = (cien_juegos / num_simulaciones) *100

print("La probabilidad de ganar jugando cien veces es:",probabilidad_de_ganar_cien_juegos,"%")
```

## 2. (25 puntos)
El juego de dados requiere que el jugador arroje dos dados una o más veces hasta que se llegue a una decisión sobre si él (o ella) gana o pierde. Él gana si la primera tirada resulta en una suma de 7 u 11 o, alternativamente, si la primera suma es 4, 5, 6, 8, 9 o 10 y la misma suma reaparece antes de que aparezca una suma de 7. Por el contrario, pierde si la primera tirada resulta en una suma de 2, 3 o 12 o, alternativamente, si la primera suma es 4, 5, 6, 8, 9 o 10 y aparece una suma de 7 antes de la primera suma reaparece.
1. Calcule la probabilidad de ganar corriendo 10,000 simulaciones.
2. Imagina que cada juego ganas o pierdes $\$50$. Si inicias con $\$500$. ¿Cuál es tu valor esperado de dinero después de 10 juegos? ¿y el más probable? ¿Cuál es la probabilidad de ganar dinero?.

```python

def dados():
    dado1 = np.random.randint(1, 7)
    dado2 = np.random.randint(1, 7)
    return dado1 + dado2

def ganar():
    suma = dados()
    if suma in [7, 11]:
        return True
    elif suma in [2, 3, 12]:
        return False
    else:
        while True:
            nueva_suma = dados()
            if nueva_suma == suma:
                return True
            elif nueva_suma == 7:
                return False

ganadas = sum(ganar() for i in range(num_simulaciones))
probabilidad_ganar = ganadas / num_simulaciones

print(f"La probabilidad de ganar es aproximadamente: {probabilidad_ganar}")

dinero_actual = 500
n = 10
Utilidad = []


for i in range(n):
    if ganar():
        dinero_actual += 50
    else:
        dinero_actual -= 50
    Utilidad.append(dinero_actual)
    
valor_esperado = np.mean(Utilidad)

valor_mas_probable = np.median(Utilidad)

probabilidad_ganar_dinero = sum(1 for dinero in Utilidad if dinero > 500) / num_simulaciones

print("Valor esperado de dinero después de 10 juegos:", valor_esperado)
print("Valor más probable de dinero después de 10 juegos:", valor_mas_probable)
print("Probabilidad de ganar dinero:", probabilidad_ganar_dinero)

```

## 3. (25 puntos)

Se eligen tres puntos uniformemente al azar del perímetro de un círculo unitario. Utilice la simulación de Monte Carlo para calcular la probabilidad de que los puntos formen un triángulo agudo, ¿cuál es la probabilidad de generar aleatoriamente triángulos agudos?

Suponga que 10.000 simulaciones son suficientes para calcular $P$.

> Ayuda: 
> 1. https://www.universoformulas.com/matematicas/trigonometria/resolucion-triangulos/
> 2. Distancia entre un par de puntos en el eje cartesiano $$d=\sqrt{(x_2-x_1)^2 + (y_2 - y_1)^2}$$
> 3. https://es.wikipedia.org/wiki/Tri%C3%A1ngulos_agudos_y_obtusos
> 4. Ecuación de un círculo unitario $x^2 + y ^2 = 1$

```python

def es_triangulo_agudo(x1, y1, x2, y2, x3, y3):
    a2 = (x2-x1)**2 + (y2-y1)**2
    b2 = (x3-x2)**2 + (y3-y2)**2
    c2 = (x1-x3)**2 + (y1-y3)**2
    return max(a2, b2, c2) < a2 + b2 + c2 - max(a2, b2, c2)

def simulacion_monte_carlo(num_simulaciones):
    num_agudos = 0

    for i in range(num_simulaciones):
        angulos = [random.uniform(0, 2*math.pi) for _ in range(3)]
        puntos = [(math.cos(angulo), math.sin(angulo)) for angulo in angulos]

        if es_triangulo_agudo(*puntos[0], *puntos[1], *puntos[2]):
            num_agudos += 1
    return num_agudos / num_simulaciones

probabilidad = simulacion_monte_carlo(num_simulaciones)
print(f"La probabilidad de que los puntos formen un triángulo agudo es aproximadamente {probabilidad}")
```

#### INTERPRETACION 3:
1. Función es_triangulo_agudo(x1, y1, x2, y2, x3, y3): Esta función determina si un triángulo formado por tres puntos en el plano cartesiano es agudo.

2. Función simulacion_monte_carlo(num_simulaciones): Esta función realiza una simulación de Monte Carlo para estimar la probabilidad de que tres puntos aleatorios seleccionados uniformemente dentro de un círculo unitario formen un triángulo agudo.

# 4. (25 puntos)

Supongamos que estás en un programa de juegos, y se le da la opción de tres puertas: detrás de una puerta hay un premio valioso; Detrás de las otras dos, cabras. Usted elige una puerta, digamos la No. 1, y el anfitrión, que sabe lo que está detrás de las puertas, abre otra puerta, supongamos la No. 3, que tiene una cabra. Luego te dice: "¿Quieres elegir la puerta No. 2?" ¿Hay ventaja al cambiar la primera puerta elegida?. Resuelva este problema usando montecarlo y responda a la siquiente pregunta:
* ¿Cuál sería la probabilidad de ganar el premio valioso si se elige cambiar la primera puerta elegida?

```python
def jugar_monty_hall(cambiar_puerta):
    puerta_ganadora = np.random.randint(3)
    puerta_elegida = np.random.randint(3)
    puertas_disponibles = list(set(range(3)) - {puerta_elegida, puerta_ganadora})
    puerta_abierta = np.random.choice(puertas_disponibles)

    if cambiar_puerta:
        puerta_elegida = list(set(range(3)) - {puerta_elegida, puerta_abierta})[0]
    return puerta_elegida == puerta_ganadora

ganadas_cambiando = sum(jugar_monty_hall(True) for _ in range(num_simulaciones))
ganadas_sin_cambiar = sum(jugar_monty_hall(False) for _ in range(num_simulaciones))

probabilidad_cambio = ganadas_cambiando / num_simulaciones
probabilidad_sin_cambio = ganadas_sin_cambiar / num_simulaciones

print(f"Probabilidad de ganar si se cambia de puerta: {probabilidad_cambio}")
print(f"Probabilidad de ganar si no se cambia de puerta: {probabilidad_sin_cambio}")
```

#### INTERPRETACION 4:

Este ejercicio ya lo habia visto en una pelicula que no recuerdo bien su nombre pero la clave es que siempre es mejor elegir otra puerta ya que aumentas tus posibilidades de ganaer. Matematicamente tiene una explicacion. De manera simplificada:
Cuando el concursante elige una puerta al principio, hay dos posibles escenarios:

1. El premio está detrás de la puerta elegida por el concursante.
  El premio está detrás de una de las otras dos puertas que no ha elegido.
2. Después de que el presentador abre una de las otras dos puertas para revelar una cabra, la información revelada cambia la probabilidad de que el premio esté detrás de cada puerta restante.

3. Si el premio estaba originalmente detrás de la puerta elegida por el concursante, el presentador puede abrir cualquiera de las otras dos puertas que contienen cabras. En este caso, si el concursante cambia su elección a la puerta restante, perderá.

4. Sin embargo, si el premio estaba originalmente detrás de una de las otras dos puertas que el concursante no eligió, el presentador solo puede abrir la puerta que contiene la otra cabra. En este caso, si el concursante cambia su elección a la puerta restante, ganará el premio.

5. Dado que hay dos veces más probabilidades de que el premio esté detrás de una de las otras dos puertas que no ha elegido inicialmente, cambiar de puerta aumenta las probabilidades de ganar el premio.
